# Monte Carlo Method for Estimating Pi - Kyle Donald

A Monte Carlo method uses random numbers to compute a value that is not random, but is unknown.

In this project we look specifically at one Monte Carlo method: using random numbers to estimate the value of π.

Using the formula for a unit circle (x^2 + y^2 = 1), we will look at the upper right quadrant of the circle.

Two random values between 0 and 1 will be generated - we can think of these as the x and y values.

If x^2 + y^2 > 1, then the point generated is outside of the circle. For example if we generate random numbers and they were both 0.8 we would have 0.8^2 + 0.8^2 = 1.28.

If x^2 + y^2 <= 1, then the point generated is inside of the circle.

The method for computing π by this Monte Carlo method is: π ≃ 4 · (outcomes inside the quarter-circle)/(total number of outcomes)

This process is repeated several times to get further estimates of π.

A graph is produced each time the code runs to visualise how the estimations of π tend to become more accurate with more iterations, but this is not always necessarily the case. 

```python
import math
import random
import matplotlib.pyplot as plt

inside_circle = 0
outside_circle = 0
total_points = 0
x_List = list()
y_List = list()

fig, ax = plt.subplots()

plt.axhline(y=math.pi, color='black', linestyle='--')
plt.xlabel("Iterations")
plt.ylabel("Value of Pi")
plt.title("Monte Carlo")

while total_points < 100:

    x = random.uniform(0,1)
    y = random.uniform(0,1)

    coordinate = (x,y)

    print(coordinate)

    total_points = total_points + 1

    ax.set(xlim=(0, total_points),
           ylim=(0, 4))

    x_List.append(total_points)

    if x*x + y*y > 1:
        outside_circle = outside_circle + 1

    elif x*x + y*y <= 1:
        inside_circle = inside_circle + 1

    print('The number of coordinates inside the circle are:',inside_circle)
    print('The number of coordinates outside the circle are:',outside_circle)

    pi = 4 * (inside_circle / total_points)
    print('The estimation of pi is:',pi)

    y_List.append(pi)

print('The real value of pi is:',math.pi)
line = plt.plot(x_List, y_List)
plt.setp(line, linewidth=3)
plt.show()
```
